### Pregunta 3.
Defina un diccionario cuyas keys sean las macroáreas de la Pregunta 2, y cuyos values sean listas de lenguas de esa macroárea. Con este diccionario, identifique el número de lenguas anotadas por macroárea. 

```python
##Primero, definimos un diccionario vacío

macroarea_lengua = {}
```

```python
##Después, recorremos la lista "macroareas"

##Largo de la lista macroareas
n = len(lenguas)

##Recorremos la lista
for i in range(n):
    ##accedemos a las macroareas en la posición i
    key = lenguas[i]
    ##accedemos a las lenguas en la posición i
    value = macroareas[i]
    ##ahora definimos el par key:value
    macroarea_lengua[key] = value
```

```python
##Definimos otro diccionario vacío

diccionario_lengua_macroarea = {}
```

```python
## recorremos la lista de macroareas únicas
for macroareas in macroareas_unicas:
    
    ## definimos una lista vacía: aqui guardamos las lenguas asociadas a esta familia
    lista_lenguas = []
    ## recorremos todas las lenguas
    for lengua in macroarea_lengua.keys():
        ## condición
        if macroareas == macroarea_lengua[lengua]:
            ## guardamos las lenguas
            lista_lenguas += [lengua]
    diccionario_lengua_macroarea[macroareas] = lista_lenguas
```

```python
###filtramos aquellas lenguas que están entre 1 y 100 hablantes:

##primer paso: identificamos las lenguas
##definimos un diccionario vacio
lista_lenguas = []
```

```python
##rellenamos la lista mediante ciclo for
for lengua in ET_wals['name']:
    ##si no está lo agregamos
    if lengua not in lista_lenguas:
        lista_lenguas += [lengua]
##imprimimos el resultado
print(lista_lenguas)
```

```python
##Parte 2: identificamos la cantidad de hablantes
lista_población = []

##rellenamos la lista mediante ciclo for
for población in ET_wals['population']:
    ##si no está lo agregamos
    lista_población += [población]
##imprimimos el resultado
print(lista_población)
```

```python
###Construimos un diccionario que luego pasaremos a un dataframe

##definimos un diccionario vacío
diccionario_lengua_población = {}

## las dos listas tienen el mismo largo, asi que elegimos cualquiera
n = len(lista_población)

## recorremos usando índices: rellenamos el diccionario con keys y values
for i in range(n):
    ## extraemos el entero
    población = lista_población[i]
    ## extraemos el string
    lengua = lista_lenguas[i]
    ## ahora, los guardamos en el diccionario
    diccionario_lengua_población[lengua] = población
```
