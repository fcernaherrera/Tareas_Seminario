**Diccionario de números en quechua: número-quechua**

Anota un número y descubre cómo se escribe en quechua

```python
##Primero hacemos una lista de equivalencias entre las unidades de los números y los términos para ellos en quechua:
##A la izquierda la unidad y a la derecha el término en quechua.
##Dentro de estas equivalencias, cabe resaltar que están ordenadas de la siguiente forma:
##Contando de izquierda a derecha, el primer elemento es 0 y el segundo es 1. 
##Contando de derecha a izquierda, el elemento a la derecha es -1 es el último elemento.
##Como vamos a trabajar con índices, no podemos definir el diez porque son dos cifras. Esto se resuelve después.
Num_quechua=[[0,""],[1,"huk"],[2,"iskay"],[3,"kimsa"],[4,"tawa"],[5,"pichqa"],[6,"suqta"],[7,"qanchis"],[8,"pusaq"],[9,"isqun"]]

##Esto se puede trabajar también como función, pero como, de momento, solo necesitamos el traductor número-quechua...
##Entonces usaremos 'input'. Definimos 'number' como un número entero que será ingresado en el recuadro emergente.
number=int(input("Ingrese: "))
##Después, definimos los rangos de valores posibles.
##Como primera condición tomamos los números entre 0 y 10 (excepto 0 y 10).
##'uni' es la variable de la unidad y se obtendrá por medio de un string que toma un número entero considerando solamente el primer dígito.
##Dentro de uni=Num_quechua[int(str(number)[-1])][1] la lógica es la siguiente: lo que debe hacer 'uni' es sondear dos equivalencias.
##Esto es para las unidades solas (por ejemplo, 2 o 5):
##Necesitamos que 'uni' tome de 'Num_quechua' solamente el primer dígito desde la derecha (el cual es [-1]).
##[1] equivale al segundo elemento dentro del par ordenado de las equivalencias en 'Num_quechua'.
if 0<number<10:
    uni=Num_quechua[int(str(number)[-1])][1]
##Como segunda condición tomamos los números entre 10 y 100 (excepto 10 y 100)
##Siguiendo lo anterior, [-2] corresponde al segundo dígito de 'number' contado desde la derecha.
elif 10<=number<100:
    uni=Num_quechua[int(str(number)[-1])][1]
    dec=Num_quechua[int(str(number)[-2])][1]
##Como tercera condición tomamos los números entre 100 y 1000 (excepto 100 y 1000)
##Siguiendo lo anterior, [-3] corresponde al tercer dígito de 'number' contado desde la derecha.
elif 100<=number<1000:
    uni=Num_quechua[int(str(number)[-1])][1]
    dec=Num_quechua[int(str(number)[-2])][1]
    cent=Num_quechua[int(str(number)[-3])][1]
##Para los numeros con tres dígitos, le añadimos un mensaje.
else:
    print("Solamente números de tres dígitos :c")

##Hasta ahora el código solo sabe devolver el mensaje...
##Así que hay que programarlo para devolver los términos correspondientes en quechua según el número ingresado.

##Para ello, primero vamos a definir los múltiplos de 10, empezando con el 10.
if number==10:
    print("chunka")
##Para los decimales entre 10-100 sin incluir 100 definimos el rango y le añadimos la noción de decimal:
##Si es un decimal, se divide entre 10 y no tiene residuo (%10==0).
##Esto funciona devolviendo el primer número del rango según fue definido anteriormente y añadiendo 'chunka'.
elif 10<number<100 and number%10==0:
    print(str(dec)+"chunka")
##Para los decimales entre 100-1000 empezamos definiendo el 100.
elif number==100:
    print("pachak")
##Para el resto de decimales entre 100-1000 definimos el rango y le añadimos la noción de decimal.
##Añadimos una condición para que si el número en el decimal es 0 (como 200) no devuelva nada como parte del número.
elif 100<number<1000 and number%10==0:
    if int(str(number)[-2])==0:
        dec==""
##Añadimos una condición para que si el número en el decimal es 1 (como 210) devuelva 'chunka' como parte del número.
    elif int(str(number)[-2])==1:
        dec="chunka"
##Si el número en el decimal no es 1 (como en 230), debe devolver el número correspondiente seguido de ' chunka'.
##El espacio en ' chunka' sirve para que los términos en quechua no se presenten juntos. Esto lo usamos después también.
    else:
        dec=str(dec)+" chunka"
##Añadimos una condición para que si el número en la centena es 1 (como 110) devuelva 'pachak' como parte del número.
    if int(str(number)[-3])==1:
        cent="pachak"
##Si el número en la centena no es 1 (como en 210), debe devolver el número correspondiente seguido de ' pachak'.
    else:
        cent=str(cent)+"pachak "
    print(cent+dec)

##Hasta aquí el código sabe manejar solamente los múltiplos de 10... Así que hay que programar el resto.

##Para los números entre 0 y 10, debe ejecutar la función 'uni' antes definida.
elif 0<number<10:
    print(uni)
##Para los números entre 10 y 20, primero debe añadirse la regla sobre las terminaciones en vocal y consonante.
##Cuando en un número >10 la unidad termina en vocal, se debe añadir '-yuq'.
elif 10<number<20:
    if uni[-1] in ["a","e","i","o","u"]:
        uni=str(uni)+"yuq"
##Cuando la unidad no termina en vocal, se debe añadir '-niyuq'
    else:
        uni=str(uni)+"niyuq"
    print("chunka "+uni)
##Para los números entre 20 y 100 la lógica es la misma.
elif 20<number<100:
    if uni[-1] in ["a","e","i","o","u"]:
        uni=str(uni)+"yuq"
    else:
        uni=str(uni)+"niyuq"
    print(dec+"chunka "+uni)
##Para los números entre 100 y 200 la lógica es la misma con dos excepciones.
elif 100<number<200:
    if uni[-1] in ["a","e","i","o","u"]:
        uni=str(uni)+"yuq"
    else:
        uni=str(uni)+"niyuq"
##Por un lado, le añadimos que no devuelva nada si el decimal es 0 (como en 105).
    if int(str(number)[-2])==0:
        dec==""
##Por el otro, le añadimos que devuelva 'chunka' cuando el término del decimal termine en 'k' (es decir, 'huk'=1)
##Esto es necesario para que no devuelva 'hukchunka', lo cual es redundante: es como decir "uno diez".
##Solo debe devolver 'chunka'.
    elif dec[-1] in ["k"]:
        dec="chunka "
    else:
        dec=str(dec)+"chunka "
    print("pachak "+dec+uni)
##Para los números entre 200 y 1000 la lógica es la misma que la anterior.
elif 200<number<1000:
    if uni[-1] in ["a","e","i","o","u"]:
        uni=str(uni)+"yuq"
    else:
        uni=str(uni)+"niyuq"
    if int(str(number)[-2])==0:
        dec==""
    elif dec[-1] in ["k"]:
        dec="chunka "
    else:
        dec=str(dec)+"chunka "
##Así como en el rango anterior, no debe devolver 'hukpachak' porque eso es como "uno cien".
##Para ello, usamos el mismo criterio. Cuando corresponda una unidad en las centenas, solo debe devolver 'pachak'.
    if cent[-1] in ["k"]:
        cent="pachak "
    else:
        cent=str(cent)+"pachak"        
    print(cent+" "+dec+uni)
else:
    print("No sé ese número en quechua :c")
    
##FIN

##Bugs: aparentemente no hay.
```
